# The Power of Social Media:
## An Instagram hashtag's effects on the Black Lives Matter movement

### Background:
   Social media allows for an exceptional increase in outreach. Within the past three years, the Black Lives Matter movement was revitalized after the murder of George Floyd by (at the time) police officer Derek Chauvin. The brutal incident was recorded and subsequently posted to Facebook, kickstarting the outrage via one of the most popular social media platforms. In response to viewing the atrocity on social media, people unsurprisingly also voiced their thoughts on social media, and solidarity began to grow as the public demanded justice. Social media was arguably at the forefront of the resuscitated movement, as seen through the mediums that discourse was being held. 

   Not only was outreach increased, but protests were arranged and organizations developed, all with the help of social media. One prominent scenario that effectively reigned on Instagram for a short period of time was when a hashtag was going viral: #blackouttuesday. This involved Instagram users posting a plain black square alongside the hashtag, typically in the caption, as a means of showing their solidarity. Debate had sparked from this phenomenon, some claiming that it was spreading the call for justice, others asserting that it was a cheap way for one to cop out while presenting a performative bravado of activism as it took little to no effort and did not result in any real ramifications.

### Why is this relevant?
One can wonder about how social media has affected society, such as if these actions which were supposedly carried out in solidarity really made any difference in the world. Perhaps there were good changes, for example laws passed or perpetrators convicted, and perhaps there were bad changes, like performative activism eclipsing earnest efforts and the murder of a man being sensationalized, or perhaps there were no changes and the performative activism was only used as a means of surface-level action.

## Research Question:
How has the traction surrounding the Black Lives Matter movement evolved on Instagram through the usage of #BlackoutTuesday starting from May of 2020? Was the hashtag 'BlackoutTuesday', which was deemed performative, more harmful or helpful for the movement?
